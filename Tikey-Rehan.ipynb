```python
kode = input("Masukkan kode tiket: ")

match kode:
    case "REG":
        nama = "Reguler"
        harga = 10000
    case "PREM":
        nama = "Premiere"
        harga = 20000
    case "VIP":
        nama = "VIP"
        harga = 30000
    case _:
        nama = "Tidak ditemukan"
        harga = 0

print("Nama Tiket :", nama)
print("Harga      :", harga)
```
